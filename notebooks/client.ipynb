# Dades extretes

Les dades que s'utilitzen per fer les cridades al client mitjançant CURL que farem servir a continuació s'han basat en el conjunt de dades de pingüins de Palmer. 

Les **dades A** són valors ficticis generats per a aquest exemple i no representen dades reals de pingüins.

En canvi, les **dades B** són un exemple real extret del conjunt de dades de pingüins de Palmer per poder veure com funciona amb dades reals.

## Logistic Regression

### Dades A

```python
!curl --request POST "http://127.0.0.1:8000/predict_lr" \
--header "Content-Type: application/json" \
--data-raw "{\
    \"island\": \"Torgersen\",\
    \"bill_length_mm\": 38.7,\
    \"bill_depth_mm\": 18.1,\
    \"flipper_length_mm\": 188.0,\
    \"body_mass_g\": 3900.0,\
    \"sex\": \"Male\"\
}"
```

Com podem veure, les dades propostes pel model tenen una alta confiança, ja que la probabilitat és extremadament alta (0.9983).

### Dades B

```python
!curl --request POST "http://127.0.0.1:8000/predict_lr" \
--header "Content-Type: application/json" \
--data-raw "{\
    \"island\": \"Dream\",\
    \"bill_length_mm\": 48.5,\
    \"bill_depth_mm\": 17.5,\
    \"flipper_length_mm\": 191.0,\
    \"body_mass_g\": 3400.0,\
    \"sex\": \"Male\"\
}"
```

De la mateixa manera que amb les dades A, el model també mostra una alta confiança en la seva predicció per a les dades B, amb una probabilitat de 0.9967. Això indica que el model és consistent en la seva capacitat per classificar correctament els pingüins basant-se en les característiques proporcionades.
